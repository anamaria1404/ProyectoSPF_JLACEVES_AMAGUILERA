### 1.4 Nodos y variables que se decidieron simular y porqué
Explicación por nodo:
1. Sold-To Party Country

Esta variable representa el país de destino que tienen los productos de Constellation Brands. Esta es una marca ampliamente reconocida a nivel mundial y cuenta con presencia en multiples regiones. Es importante estimar en el futuro cual de todos estos paises tendran un aumento en sus tendencias de compra para poder implementar adecuadas estrategias de MKT.  


2. Brand Name

Como su nombre lo dice, Brand Name representa el nombre comercial de la bebida distribuida por Constellation Brands. No todas las marcas tienen el mismo impacto y penetración en el mercado, por lo que, es importante saber cuales son las favoritas por los consumidores y así poder tener amplias estimaciones en el futuro consumo de los clientes.


3. Delivery Date - Confirmed Date

Este nodoesta representado por la diferencia entre la fecha en la que se genero la orden y la fecha en la que se surtió esa orden. Dicho nodo, creemos que es de principal relevancia, ya que hoy en día, los tiempos de entrega al consumidor juegan un roll relevante a la hora de escoger un producto/servicio. Si logramos encontrar una tendencia, podremos así encontrar y definir estrategias para minimizar dicho margen de respuesta. 

4. Varietal/Blend Description
Varietal/Blend hace referencia a la categoría principal en la cual se coloca cada bebida. Este nodo es de vital importancia para las compañias, ya que ayuda a identificar, cual tipo de producto tendrá mayor impacto y  así aplicar estrategias para su comecialización. Hoy en día el tequila es top seller, pero nos gustaría saber que necesitan las otras bebidas para lograr el mismo impacto en el mercado.
